# Python visualization

```python
import pandas as pd
```

```python
metadata = pd.read_csv("https://raw.githubusercontent.com/datacarpentry/R-genomics/gh-pages/data/Ecoli_metadata.csv")
```

```python
metadata.info()
```

```python
%matplotlib inline
```

```python
metadata['genome_size'].plot.hist()
```

```python
metadata['genome_size'].plot.hist(title='Cool Histogram')
```

```python
metadata.plot.scatter(x='generation', y='genome_size')
```

```python
metadata.plot.scatter(x='generation', y='genome_size', color='green')
```

```python
metadata.plot.scatter(x='generation', y='genome_size', marker='+')
```

```python
import plotnine as p9
```

```python
(p9.ggplot(data=metadata,
           mapping=p9.aes(x='genome_size')
))
```

```python
(p9.ggplot(data=metadata,
           mapping=p9.aes(x='genome_size'))
    + p9.geom_histogram()
)
```

```python
(p9.ggplot(data=metadata,
           mapping=p9.aes(x='genome_size'))
    + p9.geom_histogram(bins=15)
)
```
